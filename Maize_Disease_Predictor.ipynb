```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
import numpy as np
import os

# Path to the saved model
model_path = '/content/drive/MyDrive/a1.keras'

# Load the saved model
model = load_model(model_path)

# Directory containing multiple images
images_dir = '/content/drive/MyDrive/dataset/data/Blight/'  # Adjust this to your dataset directory

# List files in the directory
image_files = os.listdir(images_dir)

# Prepare to collect predictions
predictions_list = []

# Define target size for image resizing
target_size = (256, 256)

# Iterate through each image file
for image_file in image_files:
    # Construct the full path to the image
    image_path = os.path.join(images_dir, image_file)

    # Load and preprocess the image
    img = image.load_img(image_path, target_size=target_size)
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Expand dimensions to create a batch of size 1
    img_array = img_array / 255.0  # Normalize pixel values

    # Collect the preprocessed image batch
    predictions_list.append(img_array)

# Stack all images into a single numpy array
img_batch = np.vstack(predictions_list)

# Use the loaded model to make predictions on the entire batch
predictions = model.predict(img_batch)

# Get the predicted class indices for each image in the batch
predicted_class_indices = np.argmax(predictions, axis=1)

# Print the predicted class indices for each image
for i, image_file in enumerate(image_files):
    print(f" Predicted class index: {predicted_class_indices[i]}")
```
